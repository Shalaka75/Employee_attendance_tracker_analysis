#### Missing and Null value

```python
df.isnull().sum()
```

```python
df['CheckIn'] = pd.to_numeric(df['CheckIn'], errors='coerce')
df['CheckIn'].fillna(df['CheckIn'].mean(), inplace=True)


```

```python
print(df['CheckIn'].head(10))
print(df['CheckIn'].dtype)
```

```python
# Fill missing CheckIn/CheckOut with forward fill
df['CheckIn'] = df['CheckIn'].fillna(method='ffill')
df['CheckOut'] = df['CheckOut'].fillna(method='ffill')

# Fill missing Leave_Type with default category
df['Leave_Type'] = df['Leave_Type'].fillna("No Leave")

# Save cleaned file
cleaned_path = "employees_attendance_leave_tracker_cleaned.csv"
df.to_csv(cleaned_path, index=False)

cleaned_path

```
